## 3.2.7 Uygulama - Mülakat Sorusu Enumerate

Soru:

- divide_students fonksiyonu yazınız.
- Çift indexte yer alan öğrencileri bir listeye yazınız.
- Tek indexte yer alan öğrencileri başka bir listeye alınız.
- Fakat bu iki liste tek bir liste olarak return olsun.

```python
students = ["John", "Mark", "Vanessa", "Marian"]
```

```python
def divide_students(students):
    groups = [[], []]
    
    for index, student in enumerate(students):
        if index % 2 == 0:
            groups[0].append(student)
        else:
            groups[1].append(student)
    print(groups)
    return groups

divide_students(students)
        
```

## 3.2.8 Alternating Fonksiyonunun Enumerate İle Yazılması

Problem: Öyle bir fonksiyon yazmalıyız ki kendisine girilen string ifadelerin çift indexlerde yer alan karakterleri büyütsün tek indexlerde yer alan karakterleri küçültsün

```python
def alternating_with_enumerate(string):
    new_string = ""
    for i, letter in enumerate(string):
        if i % 2 == 0:
            new_string += letter.upper()
            
        else:
            new_string += letter.lower()
    
    print(new_string)
    
alternating_with_enumerate("hi my name is John an i am learning Python")
    
```

## 3.3.1 List Comprehensions

Birden fazla satır ve kod ile yapılabilen işlemleri kolayca istediğimiz çıktı veri yapısına göre tek satırda gerçekleştirme imkanı sağlayan yapılardır.

```python
salaries = [1000, 2000, 3000, 40000, 5000]
```

```python
def new_salary(x):
    return x * 20 / 100 + x
```

```python
for salary in salaries:
    print(new_salary(salary))
```

```python
null_list = []

for salary in salaries:
    null_list.append(new_salary(salary))
    
```

```python
null_list
```

```python
null_list = []
```

```python
for salary in salaries:
    if salary > 3000:
        null_list.append(new_salary(salary))
    else:
        null_list.append(new_salary(salary * 2))
```

```python
null_list
```

Eski yöntemlerle yaptık şimdi de list comprehensions ile yapalım

```python
[new_salary(salary * 2) if salary < 3000 else new_salary(salary) for salary in salaries]
```

```python
[salary * 2 for salary in salaries]
```

```python
[salary * 3 if salary < 3000 else salary for salary in salaries]
```

## NOT: 

* Eğer List Comrehensions da if tek başına ise yani else yoksa bu durumda if en sağa yazılır, for başa yazılır. Eğer if-else birarada kullanılacaksa bu durumda for bloğu en sağda kalır ve if-else bloğu sol tarafta kalır.

```python
students = ["John", "Mark", "Vanessa", "Marian"]

students_no = ["John", "Vanessa"]
```

```python
[student.lower() if student in students_no  else student.upper() for student in students]
```

```python
# not in yöntemi ile

[student.upper() if student not in students_no  else student.lower() for student in students]
```

## Lambda

* Bir fonksiyon tanımlama şeklidir def gibi ama bu kullan-at fonksiyondur.

```python
#apply ile:

df[["age", "age2", "age3"]].apply(lambda x: x / 10).head()
```

```python
def standart_scaler(col_name):
    return(col_name - col_name.mean()) / col_name.std()
    
```

```python
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()

df.head()
```

```python
#tüm korelasyonları pozitife çeviriyoruz
cor_matrix = df.corr().abs
```

```python
# 
upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k = 1).astype(np.bool))

drop_list = [col for col in upper_triangle_matrix.columns if any(upper_teriangle_matrix[col] > 0.90)]

cor_matrix[drop_list]
df.drop(drop_list, axis = 1)
```

```python
def high_correlated_cols(dataframe, plot = False, corr_th = 0.90):
    corr = dataframe.corr()
    corr_matrix = corr.abs()
    
    upper_triangle_matrix = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k = 1).astype(np.bool))
    
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_teriangle_matrix[col] > 0.90)]
    
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc = {"figure.figsize": (15, 15)})
        sns.heatmap(corr, cmap = "RdBu")
        plt.show()
        
    return drop_list
```

```python
high_correlated_cols(df)
drop_list = high_correlated_cols(df, plot = True)
df.drop(drop_list, axis=1)
high_correlated_cols(df.drop(drop_list, axis = 1), plot = True)
```
